```python
import json
import sqlite3
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

#### Connect to database that has providers, reviewers, and reviews

```python
con = sqlite3.connect('em_scraped_data_03122023_database.db')
cursor = con.cursor()
```

```python
cursor.execute("select ReviewerID, ProviderEMID from reviews")
```

```python
result = cursor.fetchall()
```

```python
relationships = []
for row in result[:1000]:
#     print(*row)
    for idx, a in enumerate(row[:-1]):
                b = row[idx + 1]
                relationships.append({"source": a, "target": b})
```

```python
relationship_df = pd.DataFrame(relationships)
```

```python
pd.set_option('display.max_rows', None)
relationship_df
```

### This cell was to make sure no relationship appears twice

```python
relationship_df["value"] = 1
relationship_df = relationship_df.groupby(["source","target"], sort=False, as_index=False).sum()
relationship_df.head(100)
```

### Graph analysis

```python
G = nx.from_pandas_edgelist(relationship_df, 
                            source = "source", 
                            target = "target", 
                            edge_attr = "value", 
                            create_using = nx.Graph())
```

```python
plt.figure(figsize=(10,10))
pos = nx.kamada_kawai_layout(G)
nx.draw(G, with_labels=True, node_color='skyblue', edge_cmap=plt.cm.Blues, pos = pos)
plt.show()
```

```python
plt.figure(figsize=(10,10))
# pos = nx.kamada_kawai_layout(G)
nx.draw(G, with_labels=True, node_color='skyblue', edge_cmap=plt.cm.Blues, pos = pos)
plt.show()
```

```python
from pyvis.network import Network
```

```python
net = Network(notebook = True, width="1000px", height="700px", bgcolor='#222222', font_color='white')

node_degree = dict(G.degree)

#Setting up node size attribute
nx.set_node_attributes(G, node_degree, 'size')

net.from_nx(G)
net.show("relationships.html")
```

```python
degree_dict = nx.degree_centrality(G)
degree_dict
```

```python
dict(G.degree)
```

```python
degree_df = pd.DataFrame.from_dict(degree_dict, orient='index', columns=['centrality'])
# Plot top 10 nodes
degree_df.sort_values('centrality', ascending=False)[0:9].plot(kind="bar")
```
